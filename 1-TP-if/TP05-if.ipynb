<h2>Fonction avec une structure conditionnelle</h2>

<h3 style="color:indigo;">Exemple</h3>

Pendant la période des soldes, un magasin décide de solder tous ses articles de moins de 100 euros à 30 % et tous ceux dont le prix est supérieur ou égal à 100 euros de 50 %.  
La fonction <span style="color:blue;">prix1</span>() définie ci-dessous retourne le prix en euros à payer pour le client en fonction du montant initial m de l'article avant les soldes.  
Cliquer dans la cellule ci-dessous, puis cliquer dans la barre d'outils sur l'onglet **Exécuter** (ou **Run** en Anglais).  
Le programme doit afficher "<b style="color:red;">49.0</b>" puis "<b style="color:red;">60.0</b>" en sortie.

```python
def prix1(m):
    if m<100:
        return 0.7*m
    else:
        return 0.5*m
#-------------------------------
print(prix1(70))
print(prix1(120))
```
